# Results of Tests

### Experiment Description

With these given tests I seek to supply answers to the question of what conditions allow for accurate pde extraction. The steps I take are thus

- Generate training data via the PDE $$u_t=-uu_x+0.02 u_{xx}$$
    - I use seeds to control the random generation of data so that I can find specific data sets with shocks
- Define a simple control experiment using the selected feature library `uu_x, u_xx, u_x_x`
    - I run training samples on a control seed (separate from seed used in experiment) to show that the hyperparameters I use during testing of shocks is indeed a suitable one
- Run the training on different time partitions of the generated data, where partitions are focused around shock or no shock
